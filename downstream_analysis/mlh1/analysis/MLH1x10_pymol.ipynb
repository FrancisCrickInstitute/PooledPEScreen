### Generates txt file with maximum function scores of missense variants assayed in MLH1 x10 saturation screen at each amino acid position 

```python
import scipy as sp
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pseudo_pre_freq_filter = 0.00014
ST_editing_filter_PEmax = 5
ST_editing_filter_PEmaxdn = 25

# IMPORT dataframe
df_variant_score = pd.read_csv(f"/Users/kajbac/code/230904_S002_final_filter/new_script/results/MLH1x10/pegRNA/variant_score/pseudo_pre_freq_{pseudo_pre_freq_filter}/{ST_editing_filter_PEmax}_PEmax_{ST_editing_filter_PEmaxdn}_PEmaxdn/data_MLH1x10_variant_score_replicates.csv")
print(df_variant_score)
```

```python
df_variant_score_2r = df_variant_score.loc[df_variant_score["number_replicates"] > 1]
df_variant_score_2r = df_variant_score_2r.drop_duplicates(subset=['var_key'])
df_variant_score_2r = df_variant_score_2r.loc[df_variant_score_2r["Consequence"]== "NON_SYNONYMOUS"]
print(df_variant_score_2r)
```

```python
df_pymol = df_variant_score_2r[["GRCh38startlocation", "GRCh38endlocation", "log2_mean_variant_replicates_normalised"]]
# print(df_pymol)
df_pymol = df_pymol.sort_values(by ="log2_mean_variant_replicates_normalised", ascending = False)
df_pymol = df_pymol.drop_duplicates(subset= "GRCh38startlocation")
df_pymol = df_pymol.sort_values(by ="GRCh38startlocation", ascending = True)
df_pymol = df_pymol.loc[df_pymol["GRCh38startlocation"] > 37017505]
df_pymol = df_pymol.loc[df_pymol["GRCh38endlocation"]< 37017600]
print(df_pymol)
```

```python
AA_list = [267,267,267,268,268,268,269,269,269,270,270,270,271,271,271,272,272,272,273,273,273,274,274,274,275,275,275,276,276,276,277,277,277,278,278,278,279,279,279,280,280,280,281,281,281,282,282,282,283,283,283,284,284,284,285,285,285,286,286,286,287,287,287,288,288,288,289,289,289,290,290,290,291,291,291,292,292,292,293,293,293,294,294,294,295,295,]
variant_position_list = []
for i in range (37017514, 37017600):
    variant_position_list.append(i)
AA_dict = dict(zip(variant_position_list,AA_list))

print(AA_dict)
```

```python
df_pymol["AA"]= df_pymol["GRCh38startlocation"].map(AA_dict)
df_pymol_AA_max = df_pymol.groupby(["AA"])["log2_mean_variant_replicates_normalised"].max()
# print(df_pymol_AA_max)
df_pymol_new = pd.merge(
        df_pymol,
        df_pymol_AA_max,
        on=["AA"],
        how="left",
        validate="many_to_one",
        suffixes=("", "_AA_max"),
    )
df_pymol_new = df_pymol_new.copy().drop_duplicates(subset = ["AA"])
print(df_pymol_new["log2_mean_variant_replicates_normalised_AA_max"])
```

```python
df_pymol_final = df_pymol_new[["AA","log2_mean_variant_replicates_normalised_AA_max"]]
print(df_pymol_final)
```

```python
df_pymol_final.to_csv("MLH1x10_nsn_pymol.txt", index=False, sep="\t")
```
